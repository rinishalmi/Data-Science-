### •	Discuss the scenarios where each scaling technique is preferred and why.

#### 1. MinMax Scaling (Normalization):

When: It’s best used when you know the data distribution is not Gaussian (i.e., normal distribution) and when you want to preserve the relationships of the original data. Ideal for neural networks and clustering algorithms like K-means.

Why: MinMax scaling transforms the data to a fixed range, typically 0 to 1. This is useful when the model's parameters need bounded values to perform optimally. Neural networks, for instance, perform better with normalized data because the convergence speed during training can improve dramatically.

#### 2. StandardScaler (Standardization):

When: It’s preferred when the data follows a Gaussian distribution, especially when the algorithm assumes or benefits from such a distribution (e.g., PCA, linear regression).

Why: StandardScaler transforms the data to have a mean of 0 and a standard deviation of 1. This standardization helps algorithms that rely on the magnitude of values or the presence of outliers.

### Encoding Techniques:

```python
## one hot encoding
```

```python
df.columns
```

```python
def find_5(df):
    column_5=[]
    for i in df.columns:
        if df[i].dtype=='O' and df[i].nunique()<=5:
            column_5.append(i)
    return column_5
    
```

```python
df['education'].nunique()
```

```python
col_one=find_5(df)
```

```python
col_one
```

```python
one_hot=pd.get_dummies(df[col_one])
one_hot
```

```python
##label encoding
```

```python
def find_col(df):
    column_rest=[]
    for i in df.columns:
        if df[i].dtype=='O' and df[i].nunique()>=5:
            column_rest.append(i)
    return column_rest
```

```python
col_label=find_col(df)
```

```python
col_label
```

```python
label_encoder=LabelEncoder()
```

```python
encoded_columns={}
for i in col_label:
        encoded_columns[i]=label_encoder.fit_transform(df[i])
encoded_df=pd.DataFrame(encoded_columns)
```

```python
encoded_df
```

### pros and cons of One-Hot Encoding and Label Encoding.

#### 1. One hot encoding
##### Pros:
* No ordinal relationship
* Model compatibility
* Interpretability
##### Cons:
* High Dimensionality
* Sparse Matrices
* Overfitting

#### 2. Label Encoding 
##### Pros:
* Simplicity
* Efficiency
* Compatibility with Ordinal Data
##### Cons:
* Misleading Relationships
* Limited Usefulness
* Less Interpretability

### Feature Selection:

```python
## using Isolation forest algorithm to identify and Remove outliers
```

```python
iso_forest=IsolationForest(contamination=0.01)
```

```python
iso_forest.fit(df_new)
```

```python
df['anomaly']=iso_forest.predict(df_new)
```

```python
outliers=df[df['anomaly']== -1]
```

```python
cleaned_df=df[df['anomaly']==1]
```
